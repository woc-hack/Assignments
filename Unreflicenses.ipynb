```python
import os
import re
```

PARSING THE FILE

```python
new_list1=[]  #TEXT
file_name1=[]  #FILE NAME
for root, dirs, files in os.walk('unreflicenses'):
    for file in files:
        file_name1.append(file)
        with open(os.path.join(root,file),'r') as f:
            text=f.read()
            new_list1.append(re.split(r'\t|\n',text))
        
        
```

```python
lst=[]
for sub_lst in new_list1:
    new_sub_list = [x for x in sub_lst if x != '']
    lst.append(new_sub_list)
    
```

EXTRACTING THE 'YOU MUST Forward Copyright notices','YOU MUST Forward OR Provide Copyright notices',
FROM EACH LICENSE FROM UNREFLICENSE AND ASSIGNING THE LABEL 

```python
lst2=[]
dict_={}
count=0
for i in range(len(lst)):
    if len(lst[i])>0:
        for j in range(len(lst[i])):
            if lst[i][j]=='YOU MUST Forward Copyright notices' or lst[i][j]=='YOU MUST Forward OR Provide Copyright notices' :
#                 print(lst[i][j])
                lst2.append(lst[i][j])
                dict_[file_name1[i]]=1
                count+=1
                break
        else:
            dict_[file_name1[i]]=0
            lst2.append('')
        
#     else:
#         lst.append('')
            
```

PARSING SPDX COLLECTION FILE

```python
new_list2 = []
file_name2=[]
for root, dirs, files in os.walk('spdx_'):
    for file in files:
        if file.endswith('.txt'):
            file_name2.append(file)
            with open(os.path.join(root, file), 'r') as f:
                text = f.read()
                new_list2.append(text)
```

FORMING DICTONARY OF FILE NAME KEY AND TEXT AS VALUE

```python
dict1={}
# print(len(new_list2))
# print(len(file_name2))
```

```python
for i in range(len(file_name2)):
    dict1[file_name2[i]]=new_list2[i]
```

ASSIGNING THE LABEL TO CORRESPONDING FILE NAME IN SPDX COLLECTION

```python
dict2={}
text_=[]
for i in range(len(file_name1)):
    for j in range(len(file_name2)):
        if file_name2[j][41:]==file_name1[i]:
            text_.append(dict1[file_name2[j]])
            #print(new_list2[j])
            dict2[file_name2[j]]=dict_[file_name1[i]]
```

```python
import pandas as pd
df=pd.DataFrame(list(dict2.items()))

df=df.rename(columns={0:'file_name',1:'label'})
```

```python
df['description']=text_
df
```

```python
# df.to_csv('unreflicnese.csv')
```

PREPROCESSING THE CORPUS

```python
df['description']=df['description'].apply(lambda x:re.sub('\S*@\S*\s?','',x))
df['description']=df['description'].apply(lambda x:re.sub(r'https?://\S+|www\.\S+', "", x))
df['description']=df['description'].apply(lambda x:x.lower())
punctuation='!"#$%&\'()*+,-./:;<=>?@[\\]^_{|}~'
df['description']=df['description'].apply(lambda x:"".join([i for i in x if i not in punctuation]))
df['description']=df['description'].apply(lambda x:re.sub('[0-9]+','',x))
df['description']=df['description'].apply(lambda x:re.split(r"\s+",x))
import nltk
from nltk.corpus import stopwords
stopword=stopwords.words('english')
df['description']=df['description'].apply(lambda x: [i for i in x if i not in stopword])
df['description'].head()
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    lemm_text=' '.join(lemm_text)
    return lemm_text
df['description']=df['description'].apply(lambda x:lemmatizer(x))
```

```python
x=df['description'] 
y=df['label']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(max_features=1000)
X=vectorizer.fit_transform(x)
# from sklearn.feature_extraction.text import TfidfTransformer
# tfidf_=TfidfTransformer(use_idf=True)
# x_tfidf=tfidf_.fit_transform(word_count)
# x_test=vectorizer.transform(x_test).toarray()
# from sklearn.feature_extraction.text import TfidfTransformer
# tfidf_transformer=TfidfTransformer()
# x_train
X.shape
```
